```python
import numpy as np
import librosa

def hann_window(N):
    return 0.5 * (1 - np.cos(2 * np.pi * np.arange(N) / (N - 1)))

def stft(x, fftsize=1024, hop=512):
    window = hann_window(fftsize)
    n_frames = 1 + (len(x) - fftsize) // hop
    X = np.zeros((fftsize // 2 + 1, n_frames), dtype=np.complex128)
    for t in range(n_frames):
        start = t * hop
        end = start + fftsize
        X[:, t] = np.fft.rfft(x[start:end] * window, fftsize)
    return X

def fft_frequencies(sr, n_fft):
    return np.linspace(0, sr / 2, int(n_fft / 2) + 1)


####  measure of the frequency at which the energy of a signal has decreased by a certain percentage
#      (often 85-95%) from its peak value. It is often used as a descriptor of the spectral shape of a signal.
def spectral_rolloff(audio_data, sr, fft_size=1024, hop_length=512, roll_percent=0.85):
    spectrogram = np.abs(stft(audio_data, fftsize=fft_size))

    # Computing the cumulative sum along the frequency axis
    cumulative_sum = np.cumsum(spectrogram, axis=0)

    # Finding the index where the cumulative sum exceeds the roll_percent
    total_energy = np.sum(spectrogram, axis=0)
    roll_threshold = roll_percent * total_energy
    rolloff_indices = np.argmax(cumulative_sum > roll_threshold, axis=0)

    # Converting the indices to frequencies
    frequencies = fft_frequencies(sr, fft_size)
    spectral_rolloff_points = frequencies[rolloff_indices]

    return spectral_rolloff_points


# audio_data, sr = librosa.load('blues.00000.wav')
# spectral_rolloff_points = spectral_rolloff(audio_data, sr)
# rolloff_mean = np.mean(spectral_rolloff_points)
# rolloff_var = np.var(spectral_rolloff_points)
# print("Spectral Rolloff Mean:", rolloff_mean)
# print("Spectral Rolloff Variance:", rolloff_var)
```

```python
# import numpy as np
# import librosa
# from scipy.fftpack import dct

# def compute_mfcc(audio_data,sample_rate , frame_size=2048, hop_length=512, num_bands=128, num_coeffs=20):
#     def create_mel_filter_bank(sample_rate, num_filters, frame_size, low_freq=0, high_freq=None):
#         if high_freq is None:
#             high_freq = sample_rate // 2

#         mel_low = 2595 * np.log10(1 + low_freq / 700)
#         mel_high = 2595 * np.log10(1 + high_freq / 700)
#         mel_points = np.linspace(mel_low, mel_high, num_filters + 2)
#         hz_points = 700 * (10**(mel_points / 2595) - 1)
#         bin_points = np.floor((frame_size + 1) * hz_points / sample_rate).astype(int)

#         filter_bank = np.zeros((num_filters, frame_size // 2 + 1))
#         for i in range(1, num_filters + 1):
#             filter_bank[i - 1, bin_points[i - 1]:bin_points[i]] = (np.arange(bin_points[i - 1], bin_points[i]) - bin_points[i - 1]) / (bin_points[i] - bin_points[i - 1])
#             filter_bank[i - 1, bin_points[i]:bin_points[i + 1]] = (bin_points[i + 1] - np.arange(bin_points[i], bin_points[i + 1])) / (bin_points[i + 1] - bin_points[i])

#         filter_bank /= np.sum(filter_bank, axis=1)[:, np.newaxis]

#         return filter_bank

#     # audio_data, sample_rate = librosa.load(audio_path, sr=None)
#     mel_filter_bank = create_mel_filter_bank(sample_rate, num_bands, frame_size)
#     stft_result = np.abs(librosa.stft(audio_data, n_fft=frame_size, hop_length=hop_length))
#     power_spectrogram = stft_result ** 2
#     mel_spectrum = np.dot(mel_filter_bank, power_spectrogram)
#     log_mel_spectrum = librosa.power_to_db(mel_spectrum, ref=np.max)
#     mfccs = dct(log_mel_spectrum, type=2, axis=0, norm='ortho')
#     mfccs = mfccs[:num_coeffs]
#     mfcc_mean = np.mean(mfccs, axis=1)
#     mfcc_var = np.var(mfccs, axis=1)

#     return mfcc_mean, mfcc_var

import numpy as np
import librosa
from scipy.fftpack import dct

def compute_mfcc(audio_data, sample_rate, frame_size=2048, hop_length=512, num_bands=128, num_coeffs=20):
    mel_filter_bank = librosa.filters.mel(sr=sample_rate, n_fft=frame_size, n_mels=num_bands)
    stft_result = np.abs(librosa.stft(audio_data, n_fft=frame_size, hop_length=hop_length))

    power_spectrogram = stft_result ** 2
    mel_spectrum = np.dot(mel_filter_bank, power_spectrogram)

    log_mel_spectrum = librosa.power_to_db(mel_spectrum, ref=np.max)

    mfccs = dct(log_mel_spectrum, type=2, axis=0, norm='ortho')
    mfccs = mfccs[:num_coeffs]

    mfcc_mean = np.mean(mfccs, axis=1)
    mfcc_var = np.var(mfccs, axis=1)

    return mfcc_mean, mfcc_var



# audio_path = "classical.00000.wav"
# audio_data, sample_rate = librosa.load(audio_path, sr=None)
# mfcc_mean, mfcc_var = compute_mfcc(audio_data, sample_rate)
# print("MFCC Mean:", mfcc_mean)
# print("MFCC Variance:", mfcc_var)
```

```python
import numpy as np
import librosa
from google.colab import files
import os

FRAME_SIZE = 1024
HOP_LENGTH = 512

# def rmse(signal, frame_size, hop_length):
#     rmse_values = []

#     # calculate rmse for each frame
#     for i in range(0, len(signal), hop_length):
#         rmse_current_frame = np.sqrt(np.sum(signal[i:i+frame_size]**2) / frame_size)
#         rmse_values.append(rmse_current_frame)

#     rms_mean = np.mean(rmse_values)
#     rms_var = np.var(rmse_values)

#     return rms_mean, rms_var

def extract_features(audio_path):
    audio_data, sr = librosa.load(audio_path)



    # Chroma features
    chroma_stft = librosa.feature.chroma_stft(y=audio_data, sr=sr, n_fft=2048, hop_length=512)
    chroma_mean = np.mean(chroma_stft)
    chroma_var = np.var(chroma_stft)



    # RMSE features
    rmse_features = rmse(audio_data, 1024, 512)



    # Spectral centroid

    # spectral_centroids = librosa.feature.spectral_centroid(y=audio_data, sr=sr)[0]
    # spectral_centroid_mean = np.mean(spectral_centroids)
    # spectral_centroid_var = np.var(spectral_centroids)

    # spectral_centroids = spectral_centroid(audio_data, sr)
    spectral_centroids,freq,spec = spectral_centroid(audio_data, sr)
    spectral_centroid_mean = np.mean(spectral_centroids)
    spectral_centroid_var = np.var(spectral_centroids)



    # Spectral Bandwidth
    spectral_bandwidth = librosa.feature.spectral_bandwidth(y=audio_data, sr=sr, n_fft=FRAME_SIZE, hop_length=HOP_LENGTH)[0]
    spectral_bandwidth_mean = np.mean(spectral_bandwidth)
    spectral_bandwidth_var = np.var(spectral_bandwidth)



    #Roll_off_mean , rolloff_var
    spectral_rolloffs = spectral_rolloff(audio_data, sr, fft_size=1024, hop_length=512, roll_percent=0.85)
    rolloff_mean = np.mean(spectral_rolloffs)
    rolloff_var = np.var(spectral_rolloffs)




    # Zero Crossing Rate
    zcr_values = zcr(audio_data , sr)
    zcr_mean = np.mean(zcr_values)
    zcr_var = np.var(zcr_values)



    # Harmonic mean and var
    harmonic = librosa.effects.harmonic(audio_data, hop_length=512)
    harmonic_mean = np.mean(harmonic)
    harmonic_var = np.var(harmonic)


    # # MFCCs
    # mfccs = librosa.feature.mfcc(y=audio_data, n_mfcc=20, sr=sr)
    # mfcc_mean = np.mean(mfccs, axis=1)
    # mfcc_var = np.var(mfccs, axis=1)
    mfcc_mean , mfcc_var = compute_mfcc(audio_data,sr, frame_size=2048, hop_length=512, num_bands=128, num_coeffs=20)




    # Tempo
    tempo, _ = librosa.beat.beat_track(y=audio_data, sr=sr)



    chroma_mean = np.array([chroma_mean])
    chroma_var = np.array([chroma_var])
    spectral_centroid_mean = np.array([spectral_centroid_mean])
    spectral_centroid_var = np.array([spectral_centroid_var])
    spectral_bandwidth_mean = np.array([spectral_bandwidth_mean])
    spectral_bandwidth_var = np.array([spectral_bandwidth_var])
    rolloff_mean = np.array([rolloff_mean])
    rolloff_var = np.array([rolloff_var])
    zcr_mean = np.array([zcr_mean])
    zcr_var = np.array([zcr_var])
    harmonic_mean = np.array([harmonic_mean])
    harmonic_var = np.array([harmonic_var])
    tempo = np.array([tempo]).flatten()


    # Combine all features
    all_features = np.concatenate((chroma_mean, chroma_var,rmse_features,
                                   spectral_centroid_mean, spectral_centroid_var,rolloff_mean,rolloff_var,
                                   spectral_bandwidth_mean, spectral_bandwidth_var,
                                   zcr_mean, zcr_var, harmonic_mean,harmonic_var , tempo))



    # all_features = np.concatenate([rmse_mean, rmse_var, [spectral_centroid_mean , spectral_centroid_var , spectral_bandwidth_mean , spectral_bandwidth_var , rolloff_mean , rolloff_var , zcr_mean , zcr_var , tempo]])


    for i in range(len(mfcc_mean)):
      all_features = np.concatenate([all_features, [mfcc_mean[i], mfcc_var[i]]])

    return all_features


uploaded = files.upload()
for audio_path in uploaded.keys():


    new_song_features = extract_features(audio_path)

    print("Extracted Features:", new_song_features)


```
